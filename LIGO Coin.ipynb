**1. Record your coin spinning!**

Click the code in the rectangle below, then press shift+return to evaluate.

```python
%matplotlib inline

import sounddevice as sd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal as sig

fs = 48000  # sampling rate, don't change
duration = 10  # seconds
recording = sd.rec(int(duration * fs), samplerate=fs, channels=1)
print('Recording for {} seconds...'.format(duration))
sd.wait()
print('Done!')
```

**2. Plot the raw sound.** (again, click in the cell then shift+return)

```python
ts = np.linspace(0, duration, duration*fs)
signal = np.array([y[0] for y in recording])
plt.plot(ts, signal);
plt.show();
```

**3. Play back the full recording**

```python
sd.play(10*signal, fs)
sd.wait()
```

**4. Choose** `start_t` **and** `end_t` **such that only the spinning coin signal is visible.**

```python
start_t = 2;
end_t = 4.5;

ind = np.logical_and(ts >= start_t, ts <= end_t)
s1 = signal[ind]
t1 = ts[ind]
plt.plot(t1, s1)
plt.show()

sd.play(10*s1, fs)
```

**5. Generate a smoothed version of the signal**

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

nwindow = 1000

ta = t1
sa = smooth(s1**2, nwindow)

plt.figure(figsize=(15,10));
ya = sa/np.max(sa)
plt.plot(ta, -ya, 'r');
plt.plot(ta, ya, 'r');
plt.fill_between(ta, -ya, ya, alpha=0.25, color='r')
plt.grid()
```

**6. Let's make a model of the signal.**

**We'll construct a model that looks like the following:**

$A = \left(\frac{t}{t_0}\right)^\alpha \sin\left[2 \pi f t \left(\frac{t}{t_0}\right)^\beta\right] \max\left[1 - \exp\left(\frac{t - t_{\rm ring}}{\tau_{\rm ring}}\right), 0\right]$

```python
t0 = 1.0;  # 1.0
tring = 2.0;  # 2.0
ringtau = 0.1;  # 0.1
omega = 5.0;  # 5.0
phase = 0.0;  # 0.0
alpha = 2.0;  # 2.0
beta = 0.5;  # 0.5
tmt = ta - min(ta);
yp = np.array([((t/t0)**alpha*np.sin(2*np.pi*omega*(t + phase)*(t/t0)**beta)*
                max(1 - np.exp(-(tring - t)/ringtau), 0))**2 for t in tmt]);
yp = yp/np.max(yp)

plt.figure(figsize=(15,10));
plt.plot(tmt, -yp, 'b', alpha=0.5);
plt.plot(tmt, yp, 'b', alpha=0.5);
plt.fill_between(tmt, -yp, yp, alpha=0.25, color='b');
plt.grid()
```

**7. Play with the numbers defined in the first six rows to find the best match to the coin signal:**

```python
t0 = 1.0;  # 1.0
tring = 2.3;  # 2.0
ringtau = 0.1;  # 0.1
omega = 2.0;  # 5.0
phase = 0.0;  # 0.0
alpha = 2.0;  # 2.0
beta = 0.5;  # 0.5
tmt = ta - min(ta);
yp = np.array([((t/t0)**alpha*np.sin(2*np.pi*omega*(t + phase)*(t/t0)**beta)*
                max(1 - np.exp(-(tring - t)/ringtau), 0))**2 for t in tmt]);
yp = yp/np.max(yp)

score = np.sum(abs(yp - ya)**2)*(100.0/len(tmt));
print('Score is (lower is better!): {:6.2f}'.format(score))

plt.figure(figsize=(15,10));
plt.plot(tmt, -ya, 'r');
plt.plot(tmt, ya, 'r');
plt.fill_between(tmt, -ya, ya, alpha=0.25, color='r')
plt.plot(tmt, -yp, 'b', alpha=0.5);
plt.plot(tmt, yp, 'b', alpha=0.5);
plt.fill_between(tmt, -yp, yp, alpha=0.25, color='b');
plt.grid()
```
